### Understanding the target variable and the  churners ratio

    The churn rate formula is: (Lost Customers ÷ Total Customers at the Start of Time Period) x 100. For example, if your business had 250 customers at the beginning of the month and lost 10 customers by the end, you would divide 10 by 250. The answer is 0.04. You then multiply 0.04 by 100, resulting in a 4% monthly churn rate.
    At this time I will visualise the churner ratio and try to analyze the potential reasons.

```python
churn_data["Churn"].value_counts().plot(kind="barh", figsize=(8,6))
plt.xlabel("Count", labelpad=14)
plt.ylabel("Target Variable",labelpad=14)
plt.title("Count of Target variable", y= 1.02)
plt.show()
```

```python

plt.figure(figsize=(20,20))
plt.title("Correlation Graph")
sns.heatmap(telco_data_features.corr(), cmap="Paired")
plt.show()
```
