# Models creating and testing

    I will try different algorithms and I will cross validated the results using predefinded function. To compare them i will use F1 weighted score because the classes are imbalanced. I will use the same logic for each algorithm:

#### Steps:
1) Finding the best parameters for the model <br>
2) Train the model <br>
3) Test the model <br>
4) Print and/or plot confustion matrix <br>
4) Print F1 weighter score because of the class imbalance <br>
5) Plot/or not a ROC curve <br>
6) Cross validation

```python
"""Cross validation function"""

def get_score(model, X_train,X_test,y_train,y_test):
    model.fit(X_train,y_train)
    return model.score(X_test,y_test)

```

## Logistic Regression

```python
LR= LogisticRegression()
```

```python
params = {
    "C" : [ 1, 5, 10,100],
    'fit_intercept': [True, False],
    'max_iter': [1000, 10000, 70000]}
    

grid_search = GridSearchCV(estimator = LR, param_grid=params)
grid_search.fit(X_train,y_train)
LR_classifier = grid_search.best_estimator_
print(LR_classifier)
```

```python
LR_model= grid_search.best_estimator_
```

```python
LR_model.fit(X_train,y_train)
```

```python
LR_model.coef_
```

```python
LR_model.intercept_
```

```python
LR_model.score(X_test,y_test)
```

```python
y_pred_LR= LR_model.predict(X_test)
```

```python
confusion_matrix_matrix = confusion_matrix(y_test, y_pred_LR)
print('Confusion matrix:')
print(f'True positives: {confusion_matrix_matrix[0][0]}')
print(f'False negative: {confusion_matrix_matrix[0][1]}')
print(f'True negative: {confusion_matrix_matrix[1][0]}')
print(f'False positive: {confusion_matrix_matrix[1][1]}')
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(confusion_matrix(y_test, y_pred_LR),
                annot=True,fmt = "d",linecolor="r",linewidths=10)
    
plt.title("Logistic Regression confusion matrix plot",fontsize=14)
plt.show()
```

```python
print(classification_report(y_test,y_pred_LR))
```

```python
LR_f1_wighted_score= f1_score(y_test,y_pred_LR, average="weighted")
```

```python
print(f'Weighted score: {LR_f1_wighted_score}.')
```

```python
y_pred_prob = LR_model.predict_proba(X_test)[:,1]
fpr, tpr, thresholds = metrics.roc_curve(y_test, y_pred_prob)
plt.plot([0, 1], [0, 1], 'k--' )
plt.plot(fpr, tpr, label='Logistic Regression',color = "r")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Logistic Regression ROC Curve',fontsize=16)
plt.show()
```

```python
KF5= KFold(shuffle=True, n_splits = 5, random_state=50)
```

```python
for train_index, test_index in KF5.split(X):
    X_train = X.iloc[train_index]
    X_test = X.iloc[test_index]
    y_train = y.iloc[train_index]
    y_test= y.iloc[test_index]
    
    print(get_score(LR_model,X_train,X_test,y_train,y_test))
```

## Defining the best model

```python
Decision_tree = tree_f1_wighted_score
Random_forest = RF_f1_wighted_score
SVC = svc_f1_wighted_score
KNN = KN_f1_wighted_score
Linear_regression= LR_f1_wighted_score

max = 0.5
L = [Decision_tree,Random_forest, SVC, KNN, Linear_regression]
for x in L:
    if x > max:
        max = x


if Decision_tree == max:
    print(f'The best model is Decision tree  with score {max}')
elif Random_forest == max:
    print(f'The best model is Random_forest with score {max}')
elif SVC == max:
    print(f'The best model is SVC gaussian  with score {max}')
elif KNN == max:
    print(f'The best model is KNeighbors  with score {max}')
else:
    print(f'The best model is Logistic regression  with score {max}')
    
```
